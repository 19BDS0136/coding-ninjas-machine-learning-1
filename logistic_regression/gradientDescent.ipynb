```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import preprocessing
from sklearn import model_selection
```

```python
cancer = datasets.load_breast_cancer()
```

```python
x = cancer.data
y = cancer.target
```

```python
xTrain,xTest,yTrain,yTest = model_selection.train_test_split(x,y)
```

```python
def cost(x,y,m):
    cost = 0
    N = len(x)
    for i in range(N):
        cost+=(1/N)*(-(y[i]*(x[i]*m).sum())+np.log(1+np.exp(((x[i]*m).sum()))))
    return cost
```

```python
def step(x,y,learnRate,m):
    N = len(x)
    M = x.shape[1]
    for j in range(M):
        slope_m = 0;
        for i in range(N):
            num = 1
            helper = -((x[i]*m).sum())
            den = 1+np.exp(helper)
            h = 1/den
            slope_m +=(-1/N)*(y[i]-h)*x[i][j]
        mNew = m[j]-(learnRate*slope_m)
        m[j]=mNew
    return m
```

```python
def gd(x,y,numIter,learnRate):
    m = np.zeros(x.shape[1])
    for i in range(numIter):
        m = step(x,y,learnRate,m)
        print(cost(x,y,m))
    return m
```

```python
def fit(x,y):
    numIter = 1000
    learnRate = 0.00001
    m = gd(x,y,numIter,learnRate)
    return m
```

```python
mFit = fit(xTrain,yTrain)
```

```python
def predict(x,m):
    N = len(x)
    y = np.zeros(N)
    for i in range(N):
        num = 1
        helper = -((x[i]*m).sum())
        den = 1+np.exp(helper)
        y[i] = 1/den
        if y[i]>0.5:
            y[i]=1
        else:
            y[i]=0
    return y
```

```python
yPre = predict(xTest,mFit)
```
