```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.plot(x,y)
plt.grid()
plt.xlabel("x")
plt.ylabel("2x")
plt.title("y=2x")
plt.show()
```

```python
plt.plot(x,y,"r+")
```

```python
#Plotting x^3

x = np.arange(0,5,0.1)
print(x.shape)
y3 = x**3
y2= x**2
plt.plot(x,y2,color="orange",linewidth=2,label="x^2")
plt.plot(x,y3,color="red",linewidth=2,label="x^3")
plt.xlabel("x")
plt.ylabel("x^3")
plt.title("y=x^3")
plt.grid()
plt.legend()
plt.savefig('comparative-plots.png')
plt.show()
```

### Pie-graph

```python
sizes = [3,4,6,2]
colors=["Blue","Purple","Red","Yellow"]
labels=["A","B","C","D"]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%")
plt.axis("equal")
plt.title("Split among classes")
plt.savefig("Pie-graph")
plt.show()
```

```python
explode=[0.1,0,0,0]
plt.pie(sizes,colors=colors,labels=labels,autopct="%.2f%%",explode=explode)
plt.axis("equal")
plt.title("Split among classes")
plt.savefig("Exploded Pie-graph")
plt.show()
```
