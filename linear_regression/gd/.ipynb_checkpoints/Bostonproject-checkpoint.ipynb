```python
import numpy as np
import pandas as pd
from sklearn import model_selection
```

```python
data = np.genfromtxt("xyTrain.csv",delimiter=",")
data.shape
```

```python
x = data[:,0:13]
y = data[:,13]
```

```python
df = pd.DataFrame(x)
```

```python
df["new"] = 1
```

```python
x = df.values
```

```python
xTrain,xTest,yTrain,yTest = model_selection.train_test_split(x,y,random_state=0)
```

```python
def step(x,y,learnRate,m):
    N = len(x)
    M = len(m)
    for i in range(M):
        slope_m = 0
        for j in range(N):
            slope_m +=(-2/N)*(y[j]-(m*x[j]).sum())*x[j][i]
        m[i] = m[i]-(learnRate*slope_m)
    return m
```

```python
def cost(x,y,m):
    N = len(x)
    ct = 0
    for i in range(N):
        ct+=(1/N)*((y[i]-((m*x[i]).sum()))**2)
    return ct
```

```python
def gd(x,y,numIter,learnRate):
    m = np.zeros(x.shape[1])
    ct = cost(x,y,m)
    for i in range(numIter):
        m  = step(x,y,learnRate,m)
        cNew = cost(x,y,m)
        print(cNew)
    return m
```

```python
def run(x,y):
    numIter = 150
    learnRate = 0.2
    m = gd(x,y,numIter,learnRate)
    dfm = pd.DataFrame(m)
    return m
```

```python
def fit(x,y):
    m = run(xTrain,yTrain)
    return m
```

```python
def pre(x,m):
    N = len(x)
    yPre = np.zeros(N)
    for i in range(N):
        yPre[i] = (x[i]*m).sum()
    return yPre
```

```python
m = fit(xTrain,yTrain)
yPre = pre(xTest,m)
```
