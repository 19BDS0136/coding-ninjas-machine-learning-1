```python
import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn import preprocessing
```

```python
data = np.genfromtxt("train.csv",delimiter=",")
```

```python
x = data[:,0:4]
y = data[:,4]
```

```python
df = pd.DataFrame(x)
df["c"] = 1
df.head()
x = df.values
x.shape
```

```python
scalar = preprocessing.StandardScaler()
scalar.fit(x)
scalar.transform(x)
```

```python
xTrain,xTest,yTrain,yTest = model_selection.train_test_split(x,y)
xTrain.shape
```

```python
def cost(x,y,m):
    cost = 0
    N = len(x)
    for i in range(N):
        cost+=(1/N)*((y[i]-(m*x[i]).sum())**2)
    return cost

def step(x,y,learnRate,m):
    N = len(x)
    M = x.shape[1]
    for j in range(M):
        slope_m = 0;
        for i in range(N):
            slope_m +=(-2/N)*(y[i]-(m*x[i]).sum())*(x[i][j])
        mNew = m[j]-(learnRate*slope_m)
        m[j]=mNew
    return m

def gd(x,y,numIter,learnRate):
    m = np.zeros(x.shape[1])
    for i in range(numIter):
        m = step(x,y,learnRate,m)
        print(cost(x,y,m))
    return m

def fit(x,y):
    numIter = 1000
    learnRate = 0.00000095
    m = gd(x,y,numIter,learnRate)
    return m
```

```python
mFit = fit(xTrain,yTrain)
```

```python
def predict(x,m):
    N = len(x)
    y = np.zeros(N)
    for i in range(N):
        y[i] = (m*x[i]).sum()
    return y
```

```python
yPre = predict(xTest,mFit)
for i in range(len(yPre)):
    yPre[i] = float("{0:.1f}".format(yPre[i]))
```

```python
xProTest = np.genfromtxt("xTest.csv",delimiter=",")
```
