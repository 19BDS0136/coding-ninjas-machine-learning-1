```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.show()
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,"y-")
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = x**3
plt.plot(x,y)
plt.show()
```

```python
z = [3,4,5,6]
labels = ["A","B","C","D"]
explode = [0.1,0,0,0]
plt.pie(z,labels=labels,explode = explode,autopct="%.1f%%",counterclock=True,startangle=90)
plt.title("division among classes")
plt.axis("equal")

plt.show()
```
